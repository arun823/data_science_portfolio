# Matrix plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
%matplotlib inline
```

```python
tips.head()
```

```python
flights.head()
```

```python
tips.corr()
```

```python
tc = tips.corr()
```

```python
sns.heatmap(tc , annot = True,cmap = 'coolwarm')
```

```python
flights
```

```python
flights.corr()
```

```python
tf = flights.pivot_table(index = 'month',columns= 'year' ,values = 'passengers')
```

```python
sns.heatmap(tf , cmap = 'coolwarm' , linecolor = 'white', linewidth = '1')
```

```python
sns.clustermap(tf , cmap = 'coolwarm' , standard_scale= 1 , figsize=(7, 7))
```

# Grids

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
pg = sns.PairGrid(iris)
pg.map(plt.scatter)
```

```python
pg = sns.PairGrid(iris)
pg.map_diag(sns.histplot)
pg.map_upper(plt.scatter)
pg.map_lower(sns.kdeplot)
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
fg = sns.FacetGrid(data = tips,row = 'time', col = 'smoker')
fg.map(sns.histplot, 'total_bill')
```

```python
fg = sns.FacetGrid(data = tips,row = 'time', col = 'smoker')
fg.map(plt.scatter, 'total_bill' , 'tip')
```

# Regression Plots

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips,hue = 'sex',markers = ['o','v'],scatter_kws = {'s' : 120})
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips,col = 'day' ,hue = 'sex',aspect = 0.6 , height = 10,scatter_kws = {'s' : 200})
```
