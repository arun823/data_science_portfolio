```python
test_df.sort_values(by = 'YearBuilt',inplace = True, ascending = True)
test_df.YearBuilt.head(10)
```

```python
test_df
```

```python
# find numerical null values
for label , content in test_df.items():
    if pd.api.types.is_numeric_dtype(content):
        if pd.isnull(content).sum():
            print(label)
```

```python
# fill numerical columns
test_df['MasVnrArea'].fillna(test_df.MasVnrArea.median(),inplace =True)
test_df['GarageYrBlt'].fillna(test_df['YearBuilt'],inplace =True)
test_df['LotFrontage'].fillna(test_df['LotFrontage'].median(),inplace =True)
test_df['GarageArea'].fillna(0.0,inplace =True)
test_df['GarageCars'].fillna(0.0,inplace =True)
test_df['BsmtHalfBath'].fillna(0.0,inplace =True)
test_df['BsmtFullBath'].fillna(0.0,inplace =True)
test_df['TotalBsmtSF'].fillna(0.0,inplace =True)
test_df['BsmtFinSF1'].fillna(0.0,inplace =True)
test_df['BsmtFinSF2'].fillna(0.0,inplace =True)
test_df['BsmtUnfSF'].fillna(0.0,inplace =True)
```

```python
# find anu more null values numerical column
for label , content in test_df.items():
    if pd.api.types.is_numeric_dtype(content):
        if pd.isnull(content).sum():
            print(label)
```

```python
# non_numeric null values
for label , content in test_df.items():
    if pd.api.types.is_string_dtype(content):
        if pd.isnull(content).sum():
           print(label)
```

```python
# fill null values of non-numerric data type
test_df['Alley'].fillna('No alley access',inplace =True)
test_df['MasVnrType'].fillna(test_df.MasVnrType.mode()[0],inplace =True)
test_df['BsmtQual'].fillna('No Basement',inplace =True)
test_df['BsmtCond'].fillna('No Basement',inplace =True)
test_df['BsmtExposure'].fillna('No Basement',inplace =True)
test_df['BsmtFinType1'].fillna('No Basement',inplace =True)
test_df['BsmtFinType2'].fillna('No Basement',inplace =True)
test_df['Electrical'].fillna(test_df['Electrical'].mode()[0],inplace =True)
test_df['FireplaceQu'].fillna('No Fireplace',inplace =True)
test_df['GarageType'].fillna('No Garage',inplace =True)
test_df['GarageFinish'].fillna('No Garage',inplace =True)
test_df['GarageQual'].fillna('No Garage',inplace =True)
test_df['GarageCond'].fillna('No Garage',inplace =True)
test_df['PoolQC'].fillna('No Pool',inplace =True)
test_df['Fence'].fillna('No Fence',inplace =True)
test_df['MSZoning'].fillna('RL',inplace =True)
test_df['Utilities'].fillna('AllPub',inplace =True)
test_df['Exterior1st'].fillna('VinylSd',inplace =True)
test_df['Exterior2nd'].fillna('VinylSd',inplace =True)
test_df['KitchenQual'].fillna('TA',inplace =True)
test_df['Functional'].fillna('Typ',inplace =True)
test_df['SaleType'].fillna('WD',inplace =True)
test_df['MiscFeature'].fillna('None',inplace =True)
```

```python
# any more null values in non_numeric columns
for label , content in test_df.items():
    if pd.api.types.is_string_dtype(content):
        if pd.isnull(content).sum():
           print(label)
```

```python
test_df.info()
```

```python
# convert object data into catogory and then in numbers
for label , content in test_df.items():
    if not pd.api.types.is_numeric_dtype(content):
        # conert string data into category
        test_df[label] = content.astype('category').cat.as_ordered()
        
        # turn categorical data into numbers
        test_df[label] = pd.Categorical(content).codes
```
