```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
x
```

```python
y = x**2
```

```python
plt.plot(x,y,'b*-')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('graph')
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,'r')
plt.subplot(2,1,2)
plt.plot(y,x,'b*-')
plt.show()
```

# object oriented method

```python
fig = plt.figure()

axis = fig.add_axes([0,0,0.8,0.8])
axis.plot(x,y)
axis.set_xlabel('x')
axis.set_ylabel('y')
axis.set_title('Graph')
```

```python
figu = plt.figure()

axes1 = figu.add_axes([0,0,1,1])
axes2 = figu.add_axes([0.05,0.35,0.45,0.45])
axes1.plot(x,y,'r*-')
axes2.plot(y,x,'b')

```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_xlabel('x')
axes[1].plot(x,y,'b')
plt.tight_layout()




```

```python
fig = plt.figure(figsize = (4,4),dpi=100)
axes = fig.add_axes([0,0,1,1,])
axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(10,5))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig('graph.png')
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2,label = 'xsquare')
axes.plot(x,x**3,label='xcube')
axes.legend(loc = 0)
```
